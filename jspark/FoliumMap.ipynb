```python
seoul_geo="C:/jspark/pythonwork/analysis/data/seoul/skorea_municipalities_geo_simple_seoul.json"
seoul_str=json.load(open(seoul_geo,encoding="utf-8"))
```

```python
seoul_str["features"]
```

```python
latlong = []
for d in seoul_str["features"]:
    tmp = list(np.stack(d["geometry"]["coordinates"][0],axis=1).mean(axis=1))
    latlong.append(tmp)
    
latlong
```

```python
name = []
for d in seoul_str["features"]:
    name.append(d["id"])
name = pd.DataFrame(name, columns=["구"])
name
```

```python
lst = pd.DataFrame(latlong, columns=["경도","위도"])
lst
```

```python
lst = pd.concat([name,lst], axis=1)
lst
```

```python
lst2 = pd.merge(lst,mean_by_gu[["구","price"]])
lst2
```

```python
from folium import plugins
```

```python
# 구별 평균 경매 감정가

map=folium.Map(location=[37.485242, 126.981770], zoom_start=10)
map.choropleth(geo_data=seoul_str, data=mean_by_gu, key_on="feature.properties.name",
              columns=["구", "price"], fill_color="YlGnBu", 
               fill_opacity=0.7, line_opacity=0.5,
              legend_name="구별 평균 감정가")

# 지도 전체화면 플러그인
plugins.Fullscreen(position='topright', title="전체화면보기", title_cancel="전체화면끄기", force_separate_button=True).add_to(map)

# 마우스 스크롤
plugins.MousePosition().add_to(map)

map
```

```python
map = folium.Map(location=[37.497911, 127.027558], zoom_start=11)
for n in lst2.index:
    folium.Marker(location=[lst2["위도"][n], lst2["경도"][n]],
                  icon=folium.Icon(color="red",icon="star"),
                  popup=lst2["구"][n]+": "+str(lst2["price"][n])).add_to(map)
    
map
```

```python
map = folium.Map(location=[37.497911, 127.027558], zoom_start=11)

for n in lst2.index:
    folium.Circle(location=[lst2["위도"][n], lst2["경도"][n]],
                  radius=lst2["price"][n]/500000,
                  color="#993366", fill_color="#993366",
                  popup=lst2["구"][n]+": "+str(lst2["price"][n])).add_to(map)
    
map
```
