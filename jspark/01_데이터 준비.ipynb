```python
# 매각진행 중인 물건
header = "https://www.gfauction.info/auction/list.html?page={}&listnum=100&orderby=&special=&court1=&damdang1=&syear=&sno=&gamMin=0&gamMax=0&eng=12%2C15&uchal_min=&uchal_max=&sday_s=&sday_e=&lowMin=0&lowMax=0&sido1=11&gugun1=&dong1=&bunji=&sagunname=&barea_min=&barea_max=&larea_min=&larea_max=&yongdo=01&special=&addr="
home = header.format(1)
page = req.urlopen(home)
soup = BS(page, "html.parser")
items = int(soup.find("p", class_="total").find("strong").get_text().replace(",",""))
pages = range(1,int(items/100)+1+1)

df = pd.DataFrame()

for i in tqdm_notebook(pages):
    url = header.format(i)
    page = req.urlopen(url)
    soup = BS(page, "html.parser")
    body = soup.find_all("td", style="border-bottom:1px solid #D9D7D7")
    data = np.reshape(body, (-1, 6))
    
    items = []
    for d in data:
        item = []
        for i in range(0,len(d)):
            tmp = d[i].get_text()
            item.append(tmp)
        items.append(item)

    tmp_df = pd.DataFrame(items)
    df = df.append(tmp_df, ignore_index=True)
```

```python
url= "https://www.gfauction.info/auction/list.html?page=1&listnum=100&orderby=&special=&court1=&damdang1=&syear=&sno=&gamMin=0&gamMax=0&eng=12%2C15&uchal_min=&uchal_max=&sday_s=&sday_e=&lowMin=0&lowMax=0&sido1=11&gugun1=&dong1=&bunji=&sagunname=&barea_min=&barea_max=&larea_min=&larea_max=&yongdo=01&special=&addr="
page = req.urlopen(url)
soup = BS(page, "html.parser")
body = soup.find_all("td", style="border-bottom:1px solid #D9D7D7")
data = np.reshape(body, (-1, 6)) # 6개씩 나눠서 array 만듦
# 데이터 쌓을 빈 데이터프레임 생성
df = pd.DataFrame(columns=["office", "num", "case_year", "cat", "addr", "load_addr", 
                           "dong", "bunji", "detail_dong", "detail_ho", "floor","brand",
                           "build_area", "land_area", "price", "lowest", "hammer", 
                           "date", "year", "month", "day", 
                           "status", "low_per", "ham_per", "comment"])

for idx in range(0,len(data)):

# 결측치가 있으면 안 되는 컬럼
    
    office = data[idx,2].find_all("p")[0].get_text().split(" ")[0]
    num = data[idx,2].find_all("p")[0].get_text().split(" ")[1]
    case_year = num.split("-")[0]
    cat = data[idx,2].find_all("p")[1].get_text().split(" ")[0].strip("[]")
    addr = " ".join(data[idx,2].find_all("p")[1].get_text().replace("\t","").split("\r\n")[0].split(" ")[1:]).strip()
    build_area = data[idx,2].find_all("p")[2].get_text().split(" ")[1].rstrip("㎡")
    price = data[idx,3].find_all("p")[0].get_text().replace(",", "")
    lowest = data[idx,3].find_all("p")[1].get_text().replace(",", "")
    date = data[idx,4].get_text().split("(")[0]
    year = date.split("-")[0]
    month = date.split("-")[1]
    day = date.split("-")[2]
    status = data[idx,5].get_text().strip("\r\n\t").split("(")[0]
    low_per = data[idx,5].get_text().strip("\r\n\t").split("(")[1].strip("%) \r\n\t")

# 상세주소 컬럼

    addr_tmp = re.sub("(\s)+", " ", addr)
    
    dong = " ".join(addr_tmp.split(" ")[0:3])

    # 0인 경우 -> 도로명주소에서 뽑아보자
    bunji = addr_tmp.split(" ")[3].strip(",동")

    addr_tmp = " ".join(addr_tmp.split(" ")[3:])
    
    # 상세주소에서 누락된 것 도로명주소에서 뽑아보자
    if re.search("([\s|,]+)[가-힣0-9]*동",addr_tmp) == None:
        detail_dong = ""
    else:
        detail_dong = re.search("([\s|,]+)[가-힣0-9]*동",addr_tmp).group().strip(" ,")
    
    if re.search("[에이비씨0-9-]*호",addr_tmp) == None:
        detail_ho = ""
    else:
        detail_ho = re.search("[에이비씨0-9-]*호",addr_tmp).group()

    if re.search("[지하0-9]*층",addr_tmp) == None:
        floor = ""
    else:
        floor = re.search("[지하0-9]*층",addr_tmp).group()
    
    addr_tmp = re.sub("([\s|,]+)[가-힣0-9]*동|[에이비씨0-9-]*호|[지하0-9]*층|[\s|(|)]+","",addr_tmp)
    
    # 브랜드 없다면 추후 API & 주소 이용해서 구해보자
    if len(addr_tmp.split(",")) < 2:
        brand = addr_tmp.split(",")[0]
    else:
        brand = addr_tmp.split(",")[-1]

# 결측치가 있을 수 있는 컬럼
    
    # 도로명주소
    if len(data[idx,2].find_all("p")[1].get_text().replace("\t","").split("\r\n")) < 2:
        load_addr = ""
    else:
        load_addr = data[idx,2].find_all("p")[1].get_text().replace("\t","").split("\r\n")[1].strip("[도로명주소] ")

    # 상세주소에서 누락된 것 도로명주소에서 뽑아보자
    if (bunji == "0") & (load_addr != ""):
        addr_tmp = re.sub("(\s)+", " ", load_addr).split(",")[0]
        if addr_tmp.split(" ")[2][-1] not in ("로", "길"):
            bunji = addr_tmp.split(" ")[3]
    
    if (detail_dong in ("동", "")) & (load_addr != "") :
        addr_tmp = " ".join(re.sub("(\s)+", " ", load_addr).split(" ")[4:])
        if re.search("[가-힣0-9]*동",addr_tmp) != None:
            detail_dong = re.search("[가-힣0-9]*동",addr_tmp).group()
        
    if (detail_ho in ("호", "")) & (load_addr != "") :
        addr_tmp = " ".join(re.sub("(\s)+", " ", load_addr).split(" ")[4:])
        if re.search("[에이비씨0-9-]*호",addr_tmp) != None:
            detail_ho = re.search("[에이비씨0-9-]*호",addr_tmp).group()

    if (floor == "") & (load_addr != "") :
        addr_tmp = " ".join(re.sub("(\s)+", " ", load_addr).split(" ")[4:])
        if re.search("[지하0-9]*층",addr_tmp) != None:
            floor = re.search("[지하0-9]*층",addr_tmp).group()
    
    # 토지면적
    if len(data[idx,2].find_all("p")[2].get_text().split(" ")) < 6:
        land_area = ""
    else:
        land_area = data[idx,2].find_all("p")[2].get_text().split(" ")[5].rstrip("㎡")
    
    # 특수사항
    if len(data[idx,2].find_all("p")) < 4:
        comment = ""
    else:
        comment = data[idx,2].find_all("p")[3].get_text().strip(" []")
        
    # 낙찰가 -> 없는 경우 최저가 낙찰인 것으로 보임.
    if len(data[idx,3].find_all("p")) < 3:
        hammer = ""
    else:
        hammer = data[idx,3].find_all("p")[2].get_text().replace(",", "")
        
    # 낙찰비율
    if len(data[idx,5].get_text().strip("\r\n\t").split("(")) < 3:
        ham_per = ""
    else:
        ham_per = data[idx,5].get_text().strip("\r\n\t").split("(")[2].strip("%)")
        
    item = [office, num, case_year, cat, addr, load_addr, 
            dong, bunji, detail_dong, detail_ho, floor, brand, 
            build_area, land_area, price, lowest, hammer, 
            date, year, month, day, 
            status, low_per, ham_per, comment]
    
    df = df.append(pd.DataFrame([item], columns=["office", "num", "case_year", "cat", "addr", "load_addr", 
                                                 "dong", "bunji", "detail_dong", "detail_ho", "floor","brand", 
                                                 "build_area", "land_area","price", "lowest", "hammer", 
                                                 "date", "year", "month", "day", 
                                                 "status", "low_per", "ham_per", "comment"]), ignore_index=True)
    
```
