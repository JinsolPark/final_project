```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/auction/auction_final.csv')
data.head(20)
```

```python
# 토지면적 유무에 따라 0/1 분류

def land_area_na(data):
    for i in range(len(data)):
        if np.isnan(data.loc[i,'land_area']):
            data.loc[i,'apt_veri'] = 0
        else :
            data.loc[i,'apt_veri'] = 1
            
land_area_na(data)
data
```

```python
# hammer, hammper 결측치를 처리 : hammer값이 널값이면 lowest값을 채운다.
# hampe값이 널값이면, hammer/price*100값을 채운다.
# 모든 값들은 매각완료된 경우만 작업한다.

def hammer_na(data):
    for i in range(len(data)):
        if data['status'][i] in ['매각','배당종결']:
            if np.isnan(data['hammer'][i]):
                # hammer 결측치 처리
                data['hammer'][i] = data['lowest'][i]
                
                # ham_per 결측치 처리
                data['ham_per'][i] = data['hammer'][i] / data['price'][i] * 100
                
            elif np.isnan(data['ham_per'][i]):
                # ham_per 결측치 처리
                data['ham_per'][i] = data['hammer'][i] / data['price'][i] * 100

hammer_na(data)
```

```python
# 결측치 확인
print((np.isnan(data["hammer"])).value_counts())
print((np.isnan(data["ham_per"])).value_counts())
```

```python
# comment 값이 있으면 comment_val값 1, 없으면 0
def comment_na(data):
    for i in range(len(data)):
        if pd.isnull(data.loc[i,'comment']):
            data.loc[i,'comment_val'] = 0
        
        else:
            data.loc[i,'comment_val'] = 1
            
comment_na(data)
data
```

```python
# 카카오 API를 이용하여, 위도, 경도, 아파트명,주소명을 추출
# 같은 주소에 상가가 포함된 경우 아파트값만 골라서 추출

cnt_err = 0
cnt = 0

def API(addr_load_name):
    for addr in addr_load_name:
        
        url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(addr)
        for i in range(len(requests.get(url,headers=header).json()['documents'])):
            try:
                if requests.get(url,headers=header).json()['documents'][i]['category_name'] in ['부동산 > 주거시설 > 아파트','부동산 > 주거시설 > 오피스텔'] :
                    place = requests.get(url,headers=header).json()['documents'][i]
    
                    addr_load_name['brand_name',cnt] = place['place_name']
                    addr_load_name['addr_lat',cnt] = place['y']
                    addr_load_name['addr_lng',cnt] = place['x']
    
                    print(addr+'-------------->' + place.get('place_name'))
    
            # 주소가 형식에 맞지않아 발생하는 에러를 예외처리
            except IndexError:
                cnt_err += 1
                print('에러 발생 :{} 회'.fromat(cnt_err))
    
            # 혹시몰라 만들어둠
            finally:
                pass
        
API(addr_load_name)
```
